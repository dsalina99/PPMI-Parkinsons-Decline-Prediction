```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from functools import reduce
```

```python
# import patient data
age = pd.read_csv("../characteristics/Age_at_visit_02Oct2024.csv")
demo = pd.read_csv("../characteristics/Demographics_02Oct2024.csv")
fam = pd.read_csv("../characteristics/Family_History_02Oct2024.csv")
saa = pd.read_csv("../Specimen_Results/SAA_Biospecimen_Analysis_Results_02Oct2024.csv")
status = pd.read_csv("../characteristics/Participant_Status_02Oct2024.csv")
# gene data has only ~950 entries, we can decide to merge later
#gene = pd.read_csv("../characteristics/PPMI_PD_Variants_Genetic_Status_WGS_20180921.csv")

dfs= [age, demo, fam, saa, status]
```

```python
# merged patient characteristics based on "PATNO" which is unified across the database
subject_characs = reduce(lambda left, right: pd.merge(left, right, on='PATNO'), dfs)
```

```python
# THis is the patients data at multiple events/visits, identified by PATNO and corresponding EVENT_ID (EVENT_ID is the follow-up visit)
subject_characs
```
